```
import pandas as pd

train_data = pd.read_csv("/content/drive/Shared drives/Kaggle/titanic/data/train.csv")
train_data = pd.DataFrame(train_data, index=[i for i in range(891)])

train_data = train_data.drop(columns=['Name', 'PassengerId'])

train_data
```

```
columns = []

for column in train_data.columns:
  columns.append(column)

columns
```

```
train_data.isnull()    
```

```
null_column = {}

for column in columns:
  for data in train_data[column].isnull():
    if data == True:
      if column not in null_column:
        null_column[column] = 1
      else:
        null_column[column] += 1

null_column
```

```
k = 0
for data in train_data['Embarked'].isnull():  
  if data == True:
    train_data = train_data.drop(index = k)
  k += 1

train_data
```

```
train_data = train_data.replace(to_replace='male', value = 1)
train_data = train_data.replace(to_replace='female', value = 0)

train_data
```

```
embarked = {}

e = 1
for data in train_data['Embarked']:
  if data not in embarked:
    embarked[data] = e
    e += 1

embarked

for key in embarked:
  train_data = train_data.replace(to_replace=key, value = embarked[key])

train_data
```

```
cabin = {}

c = 1
for data in train_data['Cabin']:
  if data not in cabin:
    cabin[data] = c
    c+=1

cabin

for key in cabin:
  train_data = train_data.replace(to_replace=key, value = cabin[key])

train_data
```

```
train_data['Age'] = train_data['Age'].fillna(train_data.mean()['Age'])

train_data['Cabin'] = train_data['Cabin'].fillna(0)

train_data
```

```
print(max(train_data['Pclass']))
train_data['Pclass'] = train_data['Pclass']/max(train_data['Pclass'])

print(max(train_data['Age']))
train_data['Age'] = train_data['Age']/max(train_data['Age'])

print(max(train_data['Embarked']))
train_data['Embarked'] = train_data['Embarked']/max(train_data['Embarked'])

train_data
```

```
train_data.to_csv("/content/drive/Shared drives/Kaggle/titanic/data/train_data_20_7_30.csv")
```
