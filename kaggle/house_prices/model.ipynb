```
import pandas as pd
import numpy as np
import tensorflow as tf

from keras.models import Sequential, load_model
from keras.layers import Dense, BatchNormalization
from keras.callbacks import EarlyStopping
from sklearn.model_selection import StratifiedKFold
```

```
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```
train_data = pd.read_csv("/content/drive/Shared drives/Kaggle/house_prices/data/train_data.csv")

dataset = pd.DataFrame(train_data)

dataset.drop(columns=['LotFrontage', 'GarageYrBlt' ,'MasVnrArea',
 'BsmtFinSF2',
 'BsmtUnfSF',
 'TotalBsmtSF',
 'BsmtFullBath',
 'BsmtHalfBath',
 'GarageCars',
 'GarageArea'], inplace=True)

dataset
```

```
datasets = dataset.values
```

```
X = datasets[:, 0:67+1].astype(float)
Y = datasets[:, 68]

Y
```

```
n_fold = 10


skf = StratifiedKFold(n_splits = n_fold, shuffle = True, random_state=seed)
```

```
for train, test in skf.split(X, Y):
  model = Sequential()
  model.add(Dense(48, input_dim = 76, activation='relu'))
  #model.add(Dense(32, activation = 'relu'))
  model.add(Dense(24, activation='relu'))
  model.add(Dense(12, activation='relu'))
  #model.add(Dense(6, activation='relu' ))
  model.add(Dense(1))
  model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
  early_stopping_callback = EarlyStopping(monitor='accuracy', patience = 10)
  model.fit(X[train], Y[train], epochs = 200, batch_size=100, verbose=0, callbacks=[early_stopping_callback])
  print(model.evaluate(X[test], Y[test])[1])
  print(model.predict(X[test], verbose=1), Y[test])
  
```

```
model = Sequential()
model.add(Dense(128, input_dim = 68, activation='relu'))
model.add(Dense(64, activation = 'relu'))
#model.add(Dense(24, activation='relu'))
#model.add(Dense(12, activation='relu'))
model.add(Dense(64, activation='relu' ))
#model.add(Dense(32, activation = 'relu'))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics=['accuracy'])
early_stopping_callback = EarlyStopping(monitor='loss', patience = 10)
model.fit(X, Y, epochs = 2000, batch_size=100, callbacks=[early_stopping_callback])
```
