```
import pandas as pd
import numpy as np
import tensorflow as tf

from keras.models import Sequential, load_model
from keras.layers import Dense, BatchNormalization
from keras.callbacks import EarlyStopping
from sklearn.model_selection import StratifiedKFold
```

```
predict_model = load_model("/content/drive/Shared drives/Kaggle/house_prices/model/2020_8_2.model")
```

```
test_data = pd.read_csv("/content/drive/Shared drives/Kaggle/house_prices/data/test_data.csv")

dataset = pd.DataFrame(test_data)


dataset.drop(columns= [ 'BsmtFinSF2',
 'BsmtUnfSF',
 'TotalBsmtSF',
 'BsmtFullBath',
 'BsmtHalfBath',
 'GarageCars',
 'GarageArea'], inplace=True)

dataset
```

```
index = pd.read_csv("/content/drive/Shared drives/Kaggle/house_prices/data/test.csv")
index = pd.DataFrame(index)

index = index['Id']

index
```

```
dataset = dataset.values
```

```
input = dataset[:, 0: 67+1]
```

```
submission = {'Id': [], 'SalePrice': []}

for index in index:
  submission['Id'].append(index)

for price in predict_model.predict(input):
  submission['SalePrice'].append(price)

submission = pd.DataFrame(submission)

submission = submission.astype('int')
```

```
submission.to_csv("/content/drive/Shared drives/Kaggle/house_prices/data/result_20_8_02.csv", index=False)
```
