```
train = pd.read_csv("/content/drive/Shared drives/Kaggle/house_prices/data/train_data_20_8_4.csv")
train= pd.DataFrame(train)
train.drop(columns=['Id'], inplace=True)
train
```

```
dataset = train.values
```
