```
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
import keras as keras
from keras.utils import to_categorical
import re
```

```
train = pd.read_csv("/content/drive/Shared drives/Kaggle/house_prices/data/test.csv")

train = pd.DataFrame(train)

train.drop(columns=['Id'], inplace=True)

train
```

```
columns = []

for column in train.columns:
  columns.append(column)
```

```
null_column = []

for column in columns:
  for bool in train[column].isnull():
    if bool ==True:
      if column not in null_column:
        null_column.append(column)    

null_column
```

```
max_value = {}
```

#train_data 

```
object_column = []

for column in columns:
  if column == 'SalePrice':
    pass
  elif train[column].dtype in ['int64', 'float64', 'float32']:
    if column not in null_column:
      a= max(train[column])
      train[column] = train[column]/a
      max_value[column] = a
    else:
      max_value[column] = max(train[column])
      datas = []
      for data in train[column]:
        datas.append(data)
      average = sum(datas)/len(datas) 
      a = max(train[column])
      max_value[column] = a
      train[column] = train[column]/a
      train[column] = train[column].fillna(average/a)
  else:
    object_column.append(column)
  
train
```

```
null_columns = []

for column in columns:
  for bool in train[column].isnull():
    if bool == True:
      if column not in null_columns:
        null_columns.append(column)

null_columns
```

```
for column in null_columns:
  train[column] = train[column].fillna(train[column].mean)
```

```
from tqdm import tqdm

for column in tqdm(null_columns):
  replace_value = {}
  for data in train[column]:
    s_data = str(data)
    if '\n' in s_data:
      if s_data not in replace_value:
        ss_data = s_data.split('\n')[0]
        replace_value[s_data] = ss_data
  for value in replace_value:
    train.replace(to_replace=value, value = replace_value[value], inplace=True)
  train[column].replace(to_replace='', value = 0, inplace = True)
  train[column] = train[column].fillna(0)
```
