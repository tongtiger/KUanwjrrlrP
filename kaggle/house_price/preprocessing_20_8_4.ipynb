```
import pandas as pd
```

```
train = pd.read_csv("/content/drive/Shared drives/Kaggle/house_prices/data/train.csv")
train = pd.DataFrame(train)

test = pd.read_csv("/content/drive/Shared drives/Kaggle/house_prices/data/test.csv")
test = pd.DataFrame(test)
```

```
test_nullcolumn = []

for column in test.columns:
  for data in test[column].isnull():
    if data == True:
      if column not in test_nullcolumn:
        test_nullcolumn.append(column)
```

```
train_nullcolumn = {}

for column in train.columns:
  for data in train[column].isnull():
    if data == True:
      if column not in train_nullcolumn:
        train_nullcolumn[column] = 1
      else:
        train_nullcolumn[column] += 1
```

```
train.info()
```

```
dtype = {'int64': [], 'float64': [], 'object' : []}

for column in train.columns:
  if train[column].dtype == 'object':
    dtype['object'].append(column)
  elif train[column].dtype == 'int64':
    dtype['int64'].append(column)
  else:
    dtype['float64'].append(column)
```

```
dtype
```

```
max_values = {}
```

```
object_values = {}
```

```
for column in train.columns:
  if column == 'SalePrice' or column == 'Id':
    pass
  elif column not in dtype['object']:
    max_value = max(train[column])
    max_values[column] = max_value
    train[column] = train[column]/max_value
  else:
    k = 1
    for data in train[column]:
      if data not in object_values:
        object_values[data] = k
        k += 1
      else:
        pass
```

```
max_values
```

```
object_values
```

```
average_value = {}
```

```
for column in train.columns:
  if column not in dtype['object']:
    average = train[column].mean
    train[column] = train[column].fillna(average)
    average_value[column] = average
  else:
    pass
```

```
average_value
```

```
for data in object_values:
  train.replace(to_replace=data, value = object_values[data], inplace=True)

train = train.fillna(0)
```

```
for column in train.columns:
  for data in train[column].isnull():
    if data == True:
      print(column)
```

```
train.to_csv("/content/drive/Shared drives/Kaggle/house_prices/data/train_data_20_8_4.csv", index=False)
```

```
for column in test.columns:
  if column == 'SalePrice' or column == 'Id':
    pass
  elif column not in dtype['object']:
    test[column] = test[column]/max_values[column]
  else:
    pass
```

```
for column in test.columns:
  if column not in dtype['object']:
    test[column] = test[column].fillna(average_value[column])
  else:
    pass
```

```
for data in object_values:
  test.replace(to_replace=data, value = object_values[data], inplace=True)

test = test.fillna(0)
```

```
for column in test.columns:
  for data in test[column].isnull():
    if data == True:
      print(column)
```
